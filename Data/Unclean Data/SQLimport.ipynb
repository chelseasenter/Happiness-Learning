```python
import sqlalchemy
import pandas as pd
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine
from config import password
```

```python
temp_files = "Clean Data/avTempData.csv"
happiness_files = "Clean Data/happinessData.csv"
username = "postgres" 
```

```python
engine = create_engine(f'postgresql://{username}:{password}@localhost:5432/Project-2')
```

```python
Base = automap_base()
Base.prepare(engine, reflect=True)
Base.classes.keys()
```

```python
AvgTemp = Base.classes.avg_temp
```

```python
Happiness = Base.classes.happiness_data
```

```python
session = Session(engine)
```

```python
temp_data = pd.read_csv(temp_files)
happiness_data = pd.read_csv(happiness_files)
```

```python
temp_data = temp_data.rename(columns={"Country":"country"})
```

```python
happiness_data.to_sql("happiness_data", con=engine, if_exists="append", index=False)
```
